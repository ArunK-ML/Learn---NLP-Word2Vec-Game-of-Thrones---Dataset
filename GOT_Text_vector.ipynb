```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk.tokenize import word_tokenize
from gensim.utils import simple_preprocess
import zipfile
import os
import nltk

# Download the necessary NLTK data
nltk.download('punkt_tab')

# Create the data directory if it doesn't exist
if not os.path.exists("data"):
  os.makedirs("data")

# Unzip the book
with zipfile.ZipFile('/content/Game of Thomas_book.zip', 'r') as zip_ref:
    zip_ref.extractall('data')

story = []
for filename in os.listdir("data"):
  # Try reading with 'latin-1' encoding
  with open(os.path.join("data", filename), encoding='latin-1') as f:
    content = f.read()
    raw_text = content.lower()
    tokens = word_tokenize(raw_text)
    # Preprocess tokens and append to the story list
    story.append(simple_preprocess(str(tokens)))
```

# **Word2Vec Model**

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar("game")
```

```python
model.wv.doesnt_match(["game", "of", "thomas"])
```

```python
model.wv.similarity("game", "of")
```

# ***PCA***

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
X = pca.fit_transform(model.wv.get_normed_vectors())
X.shape
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:500,0], y=X[:500,1], z=X[:500,2], color=y[:500])
fig.show()
```
